```python
##drop most missing null values in BMI rows

BMIdata= mergedata.dropna(axis=0)
#BMIdata.isnull().sum()


graph_dim = (40, 20)
fig, ax = plt.subplots(figsize=graph_dim)

sns.boxplot(data=BMIdata,x='Sport',y='BMI')
plt.xticks(rotation=90)
```

```python
##range of BMI
#underweight is <18.5
#normal is 18.5-25.0
#obses  >25


BMI_Level=[]

for value in BMIdata['BMI']:
    #print(value)
    
    if (value>=18.5 and value <=25.0):
        word='Normal'
        BMI_Level.append(word)
        
    elif(value>25.0):
        word='Obses'
        BMI_Level.append(word)        
    else:
        word='Under'
        BMI_Level.append(word)

BMIdata["BMI Level"] = BMI_Level  
BMIdata.head() 


```

```python

##make function to output the year,sport and only gold medal

def findGold(year,sport,Medal='Gold'):
    
    """
    parameter: Year,Sport type
    return: Athletes who receive a gold medal according to the sport type and the year.
    """
    
    
    GoldMedal_find = GoldMedal[(GoldMedal['Year']==year) & 
                              (GoldMedal['Sport']==sport)&
                               (GoldMedal['Medal']==Medal)]
    
    if any(GoldMedal_find)== False:
        return print('No gold medals awarded')
    
    return GoldMedal_find

```
